```python
from bertviz import head_view
from transformers import *
```

```python
# Load model and retrieve attention
model_version = 'bert-base-uncased'
do_lower_case = True
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)
sentence_a = "The cat sat on the mat"
sentence_b = "The cat lay on the rug"
inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt', add_special_tokens=True)
token_type_ids = inputs['token_type_ids']
input_ids = inputs['input_ids']
print(type(input_ids))
print(input_ids)
attention1 = model(input_ids, token_type_ids=token_type_ids)[-1]
print(attention1[0].shape)
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
print(type(tokens))
print(tokens)
head_view(attention1, tokens)
```

```python
# Load model and retrieve attention
model_version = 'bert-base-multilingual-cased'
do_lower_case = False
model = BertModel.from_pretrained(model_version, output_attentions=True)
tokenizer = BertTokenizer.from_pretrained(model_version, do_lower_case=do_lower_case)

sentence_a = "그 동물은 길을 건너지 않았다."
sentence_b = "왜냐하면 그것은 너무 피곤하였기 때문이다."

inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='pt', add_special_tokens=True)
token_type_ids = inputs['token_type_ids']
input_ids = inputs['input_ids']
print(type(input_ids))
print(input_ids)
attention1 = model(input_ids, token_type_ids=token_type_ids)[-1]
print(attention1[0].shape)
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
print(type(tokens))
print(tokens)
head_view(attention1, tokens)
```

```python
import numpy as np
import torch

model = TFBertModel.from_pretrained('bert-base-multilingual-cased', output_attentions=True)
tokenizer = BertTokenizer.from_pretrained('bert-base-multilingual-cased')

sentence_a = "그 동물은 길을 건너지 않았다."
sentence_b = "왜냐하면 그것은 너무 피곤하였기 때문이다."

inputs = tokenizer.encode_plus(sentence_a, sentence_b, return_tensors='tf', add_special_tokens=True)
token_type_ids = inputs['token_type_ids']
input_ids = inputs['input_ids']

train_input_ids = input_ids.numpy()

attention = model(input_ids, token_type_ids=token_type_ids)[-1]
print(attention[0].shape)
attention = attention[0].numpy()
attention = torch.from_numpy(attention)
att = (attention,)
print(type(att))
input_id_list = list(train_input_ids[0])
print(input_id_list)

tokens = tokenizer.convert_ids_to_tokens(input_id_list)

head_view(att, tokens)
```
